```python
from google.colab import drive
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

Loading the RGB image using open cv and matplotlib libraries:

```python
drive.mount('/content/drive/')
image = cv2.imread('/content/drive/MyDrive/CA01_images/color_image.png')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('RGB Image')
```

Visualizing the R, G, and B channels separately:

```python
B, G, R = cv2.split(image)

plt.imshow(R, cmap='gray')
plt.title('R Channel')
plt.show()
print('\n')
```

```python
plt.imshow(G, cmap='gray')
plt.title('G Channel')
plt.show()
print('\n')
```

```python
plt.imshow(B, cmap='gray')
plt.title('B Channel')
plt.show()
```

As it can be seen from the above plots, by seperating the channels, we are able to view the intensity of the image in each of the colors (red, green, and blue).

### **b:**

Converting the RGB image to HSV format using the cv2 function as below:

```python
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
plt.imshow(image)
plt.title('HSV Image')
```

Visualizing the H, S, and V channels separately:

```python
H, S, V = cv2.split(hsv_image)

plt.imshow(H, cmap='gray')
plt.title('H Channel')
plt.show()
print('\n')
```

```python
plt.imshow(S, cmap='gray')
plt.title('S Channel')
plt.show()
print('\n')
```

```python
plt.imshow(V, cmap='gray')
plt.title('V Channel')
plt.show()
```

As it can be seen, the image is first converted to the HSV color space and then it is displayed in each of the 3 channels (Hue, Saturation, and Value
) separately. In the Hue channel, we see color information without any brightness or saturation information. The saturation channel displays the intensity of colors in the image, and the value channel represents the brightness of the image.

### **c:**

First, I define the Hue range between 110 and 130 as below, and apply it to the HSV image by using 'cv2.inRange' and creating a mask. Next, I apply the mask to the original image by using 'cv2.bitwise_and'. Finally I visualize the blue pixels on the original image, with the non-blue pixels displayed as black:




```python
lower_bound = np.array([110, 50, 50])
upper_bound = np.array([130, 255, 255])

blue_mask = cv2.inRange(image, lower_bound, upper_bound)
blue_pixels = cv2.bitwise_and(image, image, mask=blue_mask)
black_image = np.zeros_like(image)
blue_detected = cv2.add(black_image, blue_pixels)

plt.imshow(blue_detected)
plt.title('Blue Pixels')
plt.show()
```
